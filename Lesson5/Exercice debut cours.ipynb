##### mettre la colonne inactive à true pour tous les users dont le last_seen date d'au moins un an

```python
from datetime import datetime
```

```python
def convert_str_to_datetime(str_time):
    
    return datetime.strptime(str_time, '%Y-%m-%d %H:%M:%S')
```

```python
df_people.last_seen = df_people.last_seen.apply(convert_str_to_datetime)
```

```python
def define_inactive(df_row):
    
    time = datetime.now() - df_row.last_seen
    
    if(time.days > 365):
        return False
    else:
        return True
    
```

```python
df_people['inactive'] = df_people.apply(define_inactive, axis=1)
```

```python
df_people.head()
```

###  Avec une regex: filtrer les numéros de téléphone invalides

```python
import re
```

```python
reg = r'^(?:(?:\+|00)33[\s.-]{0,3}(?:\(0\)[\s.-]{0,3})?|0)[1-9](?:(?:[\s.-]?\d{2}){4}|\d{2}(?:[\s.-]?\d{3}){2})$'
```

```python
prog = re.compile(reg)
```

```python
test_list = [True if prog.match(str(word)) is not None else False for word in df_people['phone'] ]
```

```python
df = df_people[test_list]
```

```python
df.head()
```

-------------------------------

### 3- Ajouter une colonne indiquant si le numéro de tel correspond à un téléphone portable (06/07)

```python
df['phone_number_type'] = df.apply(lambda row: '06-07' if row.phone[0:2] in ['06', '07'] else row.phone[0:2], axis=1)
```
