#### DEBUT DE L'EXERCICE :

1)
  - la somme des "Population municipale" du département
    - la somme des "Population totale" du département

```python
population_dep = population.groupby(['variable', 'Code département', 'Code région']).sum()
```

```python
population_dep.head()
```

2)
    - la part (en pourcentage) de la population municipale par rapport à la population totale

```python
pct_pop_municipale = 100 * population_dep.loc['Population municipale'] / population_dep.loc['Population totale'] 
```

```python
test = pct_pop_municipale.reset_index().drop(columns=['Code canton', 'Code arrondissement', 'Code commune']).set_index('Code département')
```
